**Method-1**

**Create a data frame uisng list**

```python
Names=["Ramesh","Suresh","Satish"]
Age=[20,25,30]
pd.DataFrame(zip(Names,Age))
```

**Update the column labels**

```python
# Adding columns labels
Names=["Ramesh","Suresh","Satish"]
Age=[20,25,30]
pd.DataFrame(zip(Names,Age),
            columns=["Names","Age"])
```

**Update the lables of rows**

```python
# adding rows labels(index)
Names=["Ramesh","Suresh","Satish"]
Age=[20,25,30]
pd.DataFrame(zip(Names,Age),
            columns=["Names","Age"],
            index=["A","B","C"])
```

```python
import random
even_list=[]
odd_list=[]
for i in range(10):
    num=random.randint(20,50)
    if num%2==0:
        even_list.append(num)
    else:
        odd_list.append(num)
print(even_list)
print(odd_list)
```

```python
df3=pd.DataFrame()
df3
```

```python
df3["Even Numbers"]=even_list
```

```python
df3
```

```python
df3["Odd Numbers"]=odd_list
```
